```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler

df = pd.read_csv("../data/raw/customers.csv")
```

```python
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')
```

```python
df['TotalCharges'] = df['TotalCharges'].fillna(df['TotalCharges'].median())
```

```python
y = df['Churn'].map({"Yes":1, "No":0})
X = df.drop(['Churn', 'customerID'], axis=1)
```

```python
cat_cols = X.select_dtypes(include='object').columns
num_cols = X.select_dtypes(exclude='object').columns
```

```python
X = pd.get_dummies(X, columns=cat_cols, drop_first=True)
```

```python
scaler = StandardScaler()
X[num_cols] = scaler.fit_transform(X[num_cols])
```

```python
X_final = X.to_numpy()
y_final = y.to_numpy()
```

```python
print("Feature matrix:", X_final.shape)
print("Label vector:", y_final.shape)
print("Example:", X_final[0], y_final[0])
```

```python
from services.ml_api_client import MLApiClient

client = MLApiClient()

predictions = []
true_labels = []
```

```python
for i in range(100):
    features = X_final[i]
    true_label = y_final[i]
    
    result = client.predict_churn(features)
    
    predictions.append(result["prediction"])
    true_labels.append(true_label)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score

accuracy = accuracy_score(true_labels, predictions)
precision = precision_score(true_labels, predictions)
recall = recall_score(true_labels, predictions)
f1 = f1_score(true_labels, predictions)
auc = roc_auc_score(true_labels, predictions)

print("Accuracy:", accuracy)
print("Precision:", precision)
print("Recall:", recall)
print("F1-score:", f1)
print("ROC-AUC:", auc)
```
